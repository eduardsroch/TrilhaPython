```python
dataf = pd.concat([Mc_2003, Mc_2013, Mc_2023])
```

```python
dataf ['DATA'] = pd.to_datetime(dataf['DATA (YYYY-MM-DD)'])
dataf.set_index(['DATA'], inplace=True)
dataf.sort_index(inplace=True)
dataf ['Ano'] = dataf.index.year
dataf ['Mes'] = dataf.index.month
dataf ['Dia'] = dataf.index.day
dataf.set_index(['Ano', 'Mes', 'Dia'], inplace=True)
dataf
```

```python
# Dados para o mês de Janeiro de cada ano
january_data = dataf[dataf.index.get_level_values('Mes') == 1]
january_stats = january_data.groupby('Ano').agg({
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': ['mean', 'min', 'max'],
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': ['mean', 'min', 'max']
})
print(january_stats)

```

```python
# Definindo as estações do ano
def definir_estacao(mes):
    if mes in [1, 2, 3]:
        return 'Verão'
    elif mes in [4, 5, 6]:
        return 'Outono'
    elif mes in [7, 8, 9]:
        return 'Inverno'
    else:
        return 'Primavera'

dataf['Estacao'] = dataf.index.get_level_values('Mes').map(definir_estacao)

# Comparando informações sobre temperatura e precipitações médias em cada estação
estacao_stats = dataf.groupby(['Ano', 'Estacao']).agg({
    'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)': 'mean',
    'TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)': 'mean'
})
print(estacao_stats)

```
